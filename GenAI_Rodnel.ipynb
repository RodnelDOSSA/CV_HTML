```python
# Instalar bibliotecas
!pip install PyPDF2 transformers

# Importar bibliotecas
import PyPDF2
# from google.colab import files
from transformers import pipeline

# Função para extrair texto de PDF
def extrair_texto_pdf(caminho_pdf):
    with open(caminho_pdf, "rb") as arquivo:
        leitor = PyPDF2.PdfReader(arquivo)
        texto = ""
        for pagina in leitor.pages:
            texto += pagina.extract_text()
    return texto


caminho_pdf = "/content/sample_data/Comprovante_Prêmio_Trading Challenge_Rodnel Dossa.pdf"

caminho_pdf2 = "/content/sample_data/Dissertação_SICREDI.pdf"


texto = extrair_texto_pdf(caminho_pdf)

texto2 = (extrair_texto_pdf(caminho_pdf2))[:10000]


# # Configurar o modelo de question-answering

modelo_qa = pipeline("question-answering", model="TunahanGokcimen/Question-Answering-xlm-roberta-base") ### O melhor modelo llm que encontrei no hugginface https://huggingface.co/

# model = "deepset/roberta-base-squad2"
# model = "bert-large-uncased-whole-word-masking-finetuned-squad"


# Função para responder perguntas
def responder_pergunta(texto, pergunta):
    resultado = modelo_qa(question=pergunta, context=texto)
    return resultado["answer"]
```

```python
# Fazer perguntas
pergunta = "Qual a Universidade citada no documento pdf?"
resposta = responder_pergunta(texto, pergunta)
print("Resposta:", resposta)
```
